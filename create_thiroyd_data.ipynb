```python
import pandas as pd

df = pd.read_csv(
    "/Users/joaoalmeida/Downloads/thyroid+disease/thyroid0387.data", header=None
)
df = df.replace("?", pd.NA)
```

```python
df
```

```python
# Define the new column names in order
new_columns = [
    "age",
    "sex",
    "on thyroxine",
    "query on thyroxine",
    "on antithyroid medication",
    "sick",
    "pregnant",
    "thyroid surgery",
    "I131 treatment",
    "query hypothyroid",
    "query hyperthyroid",
    "lithium",
    "goitre",
    "tumor",
    "hypopituitary",
    "psych",
    "TSH measured",
    "TSH",
    "T3 measured",
    "T3",
    "TT4 measured",
    "TT4",
    "T4U measured",
    "T4U",
    "FTI measured",
    "FTI",
    "TBG measured",
    "TBG",
    "referral source",
    "patient id",
]

df.columns = new_columns
df.head()
```

```python
# Split the 'patient id' column into diagnosis and id parts
df[["diagnosis", "patient_id"]] = df["patient id"].str.extract(r"^(.*)\[(.*)\]$")
df.head()
```

```python
df["diagnosis"].value_counts()
```

```python
# Map diagnosis codes to main class groups
main_class_map = {
    "A": "hyperthyroid conditions",
    "B": "hyperthyroid conditions",
    "C": "hyperthyroid conditions",
    "D": "hyperthyroid conditions",
    "E": "hypothyroid conditions",
    "F": "hypothyroid conditions",
    "G": "hypothyroid conditions",
    "H": "hypothyroid conditions",
    "I": "binding protein",
    "J": "binding protein",
    "K": "general health",
    "L": "replacement therapy",
    "M": "replacement therapy",
    "N": "replacement therapy",
    "O": "antithyroid treatment",
    "P": "antithyroid treatment",
    "Q": "antithyroid treatment",
    "R": "miscellaneous",
    "S": "miscellaneous",
    "T": "miscellaneous",
    "-": "normal",
}

df["diagnosis_main_class"] = df["diagnosis"].map(main_class_map).fillna("multiple")
df["diagnosis_main_class"].value_counts()
```

```python
df.drop(columns=["patient id", "diagnosis", "patient_id"], inplace=True)
```

```python
df
```

```python
df.columns = df.columns.str.replace(" ", "_")
df.head()
```

```python
import os

# Create the directory if it doesn't exist
output_dir = "synthetic_proxy_eval/data/thyroid"


# Save the dataframe as CSV
output_path = os.path.join(output_dir, "thyroid.csv")
df.to_csv(output_path, index=False)

print(f"Dataset saved to {output_path}")
print(f"Shape: {df.shape}")
```
